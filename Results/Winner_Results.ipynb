```python
# import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from keras.models import load_model
import pickle
import nfl_functions as nfl

pd.set_option('display.max_rows', None)

# loading the saved models
file = open('Pickles/clf_rf_win.pkl', 'rb')
clf_rf_win = pickle.load(file)
print("Loaded clf_rf_win model :: ", clf_rf_win)
file.close()

file = open('Pickles/knn_win.pkl', 'rb')
knn_win = pickle.load(file)
print("Loaded knn_win model :: ", knn_win)
file.close()

file = open('Pickles/clf_log_reg_win.pkl', 'rb')
clf_log_reg_win = pickle.load(file)
print("Loaded clf_log_reg_win model :: ", clf_log_reg_win)
file.close()

file = open('Pickles/clf_gnb_win.pkl', 'rb')
clf_gnb_win = pickle.load(file)
print("Loaded clf_gnb_win model :: ", clf_gnb_win)
file.close()

ann_win = load_model('Pickles/ann_win.h5')
print("Loaded ann_win model :: ", ann_win)
```

```python
# load weeks 2-16 season 2020 files
w2 = pd.read_csv('Data/df_2020_w2.csv')
w3 = pd.read_csv('Data/df_2020_w3.csv')
w4 = pd.read_csv('Data/df_2020_w4.csv')
w5 = pd.read_csv('Data/df_2020_w5.csv')
w6 = pd.read_csv('Data/df_2020_w6.csv')
w7 = pd.read_csv('Data/df_2020_w7.csv')
w8 = pd.read_csv('Data/df_2020_w8.csv')
w9 = pd.read_csv('Data/df_2020_w9.csv')
w10 = pd.read_csv('Data/df_2020_w10.csv')
w11 = pd.read_csv('Data/df_2020_w11.csv')
w12 = pd.read_csv('Data/df_2020_w12.csv')
w13 = pd.read_csv('Data/df_2020_w13.csv')
w14 = pd.read_csv('Data/df_2020_w14.csv')
w15 = pd.read_csv('Data/df_2020_w15.csv')
w16 = pd.read_csv('Data/df_2020_w16.csv')
```

```python
rf = pd.concat([nfl.predict_winners(w2, clf_rf_win), nfl.predict_winners(w3, clf_rf_win),
                nfl.predict_winners(w4, clf_rf_win), nfl.predict_winners(w5, clf_rf_win),
                nfl.predict_winners(w6, clf_rf_win), nfl.predict_winners(w7, clf_rf_win),
                nfl.predict_winners(w8, clf_rf_win), nfl.predict_winners(w9, clf_rf_win),
                nfl.predict_winners(w10, clf_rf_win), nfl.predict_winners(w11, clf_rf_win),
                nfl.predict_winners(w12, clf_rf_win), nfl.predict_winners(w13, clf_rf_win),
                nfl.predict_winners(w14, clf_rf_win), nfl.predict_winners(w15, clf_rf_win),
                nfl.predict_winners(w16, clf_rf_win)]).reset_index(drop=True)

knn = pd.concat([nfl.predict_winners(w2, knn_win), nfl.predict_winners(w3, knn_win),
                 nfl.predict_winners(w4, knn_win), nfl.predict_winners(w5, knn_win),
                 nfl.predict_winners(w6, knn_win), nfl.predict_winners(w7, knn_win),
                 nfl.predict_winners(w8, knn_win), nfl.predict_winners(w9, knn_win),
                 nfl.predict_winners(w10, knn_win), nfl.predict_winners(w11, knn_win),
                 nfl.predict_winners(w12, knn_win), nfl.predict_winners(w13, knn_win),
                 nfl.predict_winners(w14, knn_win), nfl.predict_winners(w15, knn_win),
                 nfl.predict_winners(w16, knn_win)]).reset_index(drop=True)

lr = pd.concat([nfl.predict_winners(w2, clf_log_reg_win), nfl.predict_winners(w3, clf_log_reg_win),
                nfl.predict_winners(w4, clf_log_reg_win), nfl.predict_winners(w5, clf_log_reg_win),
                nfl.predict_winners(w6, clf_log_reg_win), nfl.predict_winners(w7, clf_log_reg_win),
                nfl.predict_winners(w8, clf_log_reg_win), nfl.predict_winners(w9, clf_log_reg_win),
                nfl.predict_winners(w10, clf_log_reg_win), nfl.predict_winners(w11, clf_log_reg_win),
                nfl.predict_winners(w12, clf_log_reg_win), nfl.predict_winners(w13, clf_log_reg_win),
                nfl.predict_winners(w14, clf_log_reg_win), nfl.predict_winners(w15, clf_log_reg_win),
                nfl.predict_winners(w16, clf_log_reg_win)]).reset_index(drop=True)
                    
gnb = pd.concat([nfl.predict_winners(w2, clf_gnb_win), nfl.predict_winners(w3, clf_gnb_win),
                 nfl.predict_winners(w4, clf_gnb_win), nfl.predict_winners(w5, clf_gnb_win),
                 nfl.predict_winners(w6, clf_gnb_win), nfl.predict_winners(w7, clf_gnb_win),
                 nfl.predict_winners(w8, clf_gnb_win), nfl.predict_winners(w9, clf_gnb_win),
                 nfl.predict_winners(w10, clf_gnb_win), nfl.predict_winners(w11, clf_gnb_win),
                 nfl.predict_winners(w12, clf_gnb_win), nfl.predict_winners(w13, clf_gnb_win),
                 nfl.predict_winners(w14, clf_gnb_win), nfl.predict_winners(w15, clf_gnb_win),
                 nfl.predict_winners(w16, clf_gnb_win)]).reset_index(drop=True)

ann = pd.concat([nfl.keras_winner(w2, ann_win), nfl.keras_winner(w3, ann_win),
                 nfl.keras_winner(w4, ann_win), nfl.keras_winner(w5, ann_win),
                 nfl.keras_winner(w6, ann_win), nfl.keras_winner(w7, ann_win),
                 nfl.keras_winner(w8, ann_win), nfl.keras_winner(w9, ann_win),
                 nfl.keras_winner(w10, ann_win), nfl.keras_winner(w11, ann_win),
                 nfl.keras_winner(w12, ann_win), nfl.keras_winner(w13, ann_win),
                 nfl.keras_winner(w14, ann_win), nfl.keras_winner(w15, ann_win),
                 nfl.keras_winner(w16, ann_win)]).reset_index(drop=True)
```

```python
print('RF % Correct:', round(rf.correct.mean()*100, 3))
print('KNN % Correct:', round(knn.correct.mean()*100, 3))
print('LR % Correct:', round(lr.correct.mean()*100, 3))
print('GNB % Correct:', round(gnb.correct.mean()*100, 3))
print('ANN % Correct:', round(ann.correct.mean()*100, 3))
```
