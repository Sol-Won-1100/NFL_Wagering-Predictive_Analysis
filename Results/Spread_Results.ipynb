```python
# import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from keras.models import load_model
import pickle
import nfl_functions as nfl

pd.set_option('display.max_rows', None)

# loading the saved models
file = open('Pickles/clf_rf_spread.pkl', 'rb')
clf_rf_spread = pickle.load(file)
print("Loaded clf_rf_spread model :: ", clf_rf_spread)
file.close()

file = open('Pickles/knn_spread.pkl', 'rb')
knn_spread = pickle.load(file)
print("Loaded knn_spread model :: ", knn_spread)
file.close()

file = open('Pickles/clf_log_reg_spread.pkl', 'rb')
clf_log_reg_spread = pickle.load(file)
print("Loaded clf_log_reg_spread model :: ", clf_log_reg_spread)
file.close()

file = open('Pickles/clf_gnb_spread.pkl', 'rb')
clf_gnb_spread = pickle.load(file)
print("Loaded clf_gnb_spread model :: ", clf_gnb_spread)
file.close()

ann_spread = load_model('Pickles/ann_spread.h5')
print("Loaded ann_spread model :: ", ann_spread)
```

```python
# load weeks 2-16 season 2020 files
w2 = pd.read_csv('Data/df_2020_w2.csv')
w3 = pd.read_csv('Data/df_2020_w3.csv')
w4 = pd.read_csv('Data/df_2020_w4.csv')
w5 = pd.read_csv('Data/df_2020_w5.csv')
w6 = pd.read_csv('Data/df_2020_w6.csv')
w7 = pd.read_csv('Data/df_2020_w7.csv')
w8 = pd.read_csv('Data/df_2020_w8.csv')
w9 = pd.read_csv('Data/df_2020_w9.csv')
w10 = pd.read_csv('Data/df_2020_w10.csv')
w11 = pd.read_csv('Data/df_2020_w11.csv')
w12 = pd.read_csv('Data/df_2020_w12.csv')
w13 = pd.read_csv('Data/df_2020_w13.csv')
w14 = pd.read_csv('Data/df_2020_w14.csv')
w15 = pd.read_csv('Data/df_2020_w15.csv')
w16 = pd.read_csv('Data/df_2020_w16.csv')
```

```python
rf = pd.concat([nfl.predict_spread(w2, clf_rf_spread), nfl.predict_spread(w3, clf_rf_spread),
                nfl.predict_spread(w4, clf_rf_spread), nfl.predict_spread(w5, clf_rf_spread),
                nfl.predict_spread(w6, clf_rf_spread), nfl.predict_spread(w7, clf_rf_spread),
                nfl.predict_spread(w8, clf_rf_spread), nfl.predict_spread(w9, clf_rf_spread),
                nfl.predict_spread(w10, clf_rf_spread), nfl.predict_spread(w11, clf_rf_spread),
                nfl.predict_spread(w12, clf_rf_spread), nfl.predict_spread(w13, clf_rf_spread),
                nfl.predict_spread(w14, clf_rf_spread), nfl.predict_spread(w15, clf_rf_spread),
                nfl.predict_spread(w16, clf_rf_spread)]).reset_index(drop=True)

knn = pd.concat([nfl.predict_spread(w2, knn_spread), nfl.predict_spread(w3, knn_spread),
                 nfl.predict_spread(w4, knn_spread), nfl.predict_spread(w5, knn_spread),
                 nfl.predict_spread(w6, knn_spread), nfl.predict_spread(w7, knn_spread),
                 nfl.predict_spread(w8, knn_spread), nfl.predict_spread(w9, knn_spread),
                 nfl.predict_spread(w10, knn_spread), nfl.predict_spread(w11, knn_spread),
                 nfl.predict_spread(w12, knn_spread), nfl.predict_spread(w13, knn_spread),
                 nfl.predict_spread(w14, knn_spread), nfl.predict_spread(w15, knn_spread),
                 nfl.predict_spread(w16, knn_spread)]).reset_index(drop=True)

lr = pd.concat([nfl.predict_spread(w2, clf_log_reg_spread), nfl.predict_spread(w3, clf_log_reg_spread),
                nfl.predict_spread(w4, clf_log_reg_spread), nfl.predict_spread(w5, clf_log_reg_spread),
                nfl.predict_spread(w6, clf_log_reg_spread), nfl.predict_spread(w7, clf_log_reg_spread),
                nfl.predict_spread(w8, clf_log_reg_spread), nfl.predict_spread(w9, clf_log_reg_spread),
                nfl.predict_spread(w10, clf_log_reg_spread), nfl.predict_spread(w11, clf_log_reg_spread),
                nfl.predict_spread(w12, clf_log_reg_spread), nfl.predict_spread(w13, clf_log_reg_spread),
                nfl.predict_spread(w14, clf_log_reg_spread), nfl.predict_spread(w15, clf_log_reg_spread),
                nfl.predict_spread(w16, clf_log_reg_spread)]).reset_index(drop=True)
                    
gnb = pd.concat([nfl.predict_spread(w2, clf_gnb_spread), nfl.predict_spread(w3, clf_gnb_spread),
                 nfl.predict_spread(w4, clf_gnb_spread), nfl.predict_spread(w5, clf_gnb_spread),
                 nfl.predict_spread(w6, clf_gnb_spread), nfl.predict_spread(w7, clf_gnb_spread),
                 nfl.predict_spread(w8, clf_gnb_spread), nfl.predict_spread(w9, clf_gnb_spread),
                 nfl.predict_spread(w10, clf_gnb_spread), nfl.predict_spread(w11, clf_gnb_spread),
                 nfl.predict_spread(w12, clf_gnb_spread), nfl.predict_spread(w13, clf_gnb_spread),
                 nfl.predict_spread(w14, clf_gnb_spread), nfl.predict_spread(w15, clf_gnb_spread),
                 nfl.predict_spread(w16, clf_gnb_spread)]).reset_index(drop=True)

ann = pd.concat([nfl.keras_spread(w2, ann_spread), nfl.keras_spread(w3, ann_spread),
                 nfl.keras_spread(w4, ann_spread), nfl.keras_spread(w5, ann_spread),
                 nfl.keras_spread(w6, ann_spread), nfl.keras_spread(w7, ann_spread),
                 nfl.keras_spread(w8, ann_spread), nfl.keras_spread(w9, ann_spread),
                 nfl.keras_spread(w10, ann_spread), nfl.keras_spread(w11, ann_spread),
                 nfl.keras_spread(w12, ann_spread), nfl.keras_spread(w13, ann_spread),
                 nfl.keras_spread(w14, ann_spread), nfl.keras_spread(w15, ann_spread),
                 nfl.keras_spread(w16, ann_spread)]).reset_index(drop=True)
```

```python
print('RF %:', round(rf.correct.mean()*100, 3))
print('KNN %:', round(knn.correct.mean()*100, 3))
print('LR %:', round(lr.correct.mean()*100, 3))
print('GNB %:', round(gnb.correct.mean()*100, 3))
print('ANN %:', round(ann.correct.mean()*100, 3))
```
