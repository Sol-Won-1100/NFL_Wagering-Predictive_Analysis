```python
# import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from keras.models import load_model
import pickle
import nfl_functions as nfl

pd.set_option('display.max_rows', None)

# loading the saved models
file = open('Pickles/clf_rf_ou.pkl', 'rb')
clf_rf_ou = pickle.load(file)
print("Loaded clf_rf_ou model :: ", clf_rf_ou)
file.close()

file = open('Pickles/knn_ou.pkl', 'rb')
knn_ou = pickle.load(file)
print("Loaded knn_ou model :: ", knn_ou)
file.close()

file = open('Pickles/clf_log_reg_ou.pkl', 'rb')
clf_log_reg_ou = pickle.load(file)
print("Loaded clf_log_reg_ou model :: ", clf_log_reg_ou)
file.close()

file = open('Pickles/clf_gnb_ou.pkl', 'rb')
clf_gnb_ou = pickle.load(file)
print("Loaded clf_gnb_ou model :: ", clf_gnb_ou)
file.close()

ann_ou = load_model('Pickles/ann_ou.h5')
print("Loaded ann_ou model :: ", ann_ou)
```

```python
# load weeks 2-16 season 2020 files
w2 = pd.read_csv('Data/df_2020_w2.csv')
w3 = pd.read_csv('Data/df_2020_w3.csv')
w4 = pd.read_csv('Data/df_2020_w4.csv')
w5 = pd.read_csv('Data/df_2020_w5.csv')
w6 = pd.read_csv('Data/df_2020_w6.csv')
w7 = pd.read_csv('Data/df_2020_w7.csv')
w8 = pd.read_csv('Data/df_2020_w8.csv')
w9 = pd.read_csv('Data/df_2020_w9.csv')
w10 = pd.read_csv('Data/df_2020_w10.csv')
w11 = pd.read_csv('Data/df_2020_w11.csv')
w12 = pd.read_csv('Data/df_2020_w12.csv')
w13 = pd.read_csv('Data/df_2020_w13.csv')
w14 = pd.read_csv('Data/df_2020_w14.csv')
w15 = pd.read_csv('Data/df_2020_w15.csv')
w16 = pd.read_csv('Data/df_2020_w16.csv')
```

```python
rf = pd.concat([nfl.predict_ou(w2, clf_rf_ou), nfl.predict_ou(w3, clf_rf_ou),
                nfl.predict_ou(w4, clf_rf_ou), nfl.predict_ou(w5, clf_rf_ou),
                nfl.predict_ou(w6, clf_rf_ou), nfl.predict_ou(w7, clf_rf_ou),
                nfl.predict_ou(w8, clf_rf_ou), nfl.predict_ou(w9, clf_rf_ou),
                nfl.predict_ou(w10, clf_rf_ou), nfl.predict_ou(w11, clf_rf_ou),
                nfl.predict_ou(w12, clf_rf_ou), nfl.predict_ou(w13, clf_rf_ou),
                nfl.predict_ou(w14, clf_rf_ou), nfl.predict_ou(w15, clf_rf_ou),
                nfl.predict_ou(w16, clf_rf_ou)]).reset_index(drop = True)

knn = pd.concat([nfl.predict_ou(w2, knn_ou), nfl.predict_ou(w3, knn_ou),
                 nfl.predict_ou(w4, knn_ou), nfl.predict_ou(w5, knn_ou),
                 nfl.predict_ou(w6, knn_ou), nfl.predict_ou(w7, knn_ou),
                 nfl.predict_ou(w8, knn_ou), nfl.predict_ou(w9, knn_ou),
                 nfl.predict_ou(w10, knn_ou), nfl.predict_ou(w11, knn_ou),
                 nfl.predict_ou(w12, knn_ou), nfl.predict_ou(w13, knn_ou),
                 nfl.predict_ou(w14, knn_ou), nfl.predict_ou(w15, knn_ou),
                 nfl.predict_ou(w16, knn_ou)]).reset_index(drop = True)

lr = pd.concat([nfl.predict_ou(w2, clf_log_reg_ou), nfl.predict_ou(w3, clf_log_reg_ou),
                nfl.predict_ou(w4, clf_log_reg_ou), nfl.predict_ou(w5, clf_log_reg_ou),
                nfl.predict_ou(w6, clf_log_reg_ou), nfl.predict_ou(w7, clf_log_reg_ou),
                nfl.predict_ou(w8, clf_log_reg_ou), nfl.predict_ou(w9, clf_log_reg_ou),
                nfl.predict_ou(w10, clf_log_reg_ou), nfl.predict_ou(w11, clf_log_reg_ou),
                nfl.predict_ou(w12, clf_log_reg_ou), nfl.predict_ou(w13, clf_log_reg_ou),
                nfl.predict_ou(w14, clf_log_reg_ou), nfl.predict_ou(w15, clf_log_reg_ou),
                nfl.predict_ou(w16, clf_log_reg_ou)]).reset_index(drop = True)
                    
gnb = pd.concat([nfl.predict_ou(w2, clf_gnb_ou), nfl.predict_ou(w3, clf_gnb_ou),
                 nfl.predict_ou(w4, clf_gnb_ou), nfl.predict_ou(w5, clf_gnb_ou),
                 nfl.predict_ou(w6, clf_gnb_ou), nfl.predict_ou(w7, clf_gnb_ou),
                 nfl.predict_ou(w8, clf_gnb_ou), nfl.predict_ou(w9, clf_gnb_ou),
                 nfl.predict_ou(w10, clf_gnb_ou), nfl.predict_ou(w11, clf_gnb_ou),
                 nfl.predict_ou(w12, clf_gnb_ou), nfl.predict_ou(w13, clf_gnb_ou),
                 nfl.predict_ou(w14, clf_gnb_ou), nfl.predict_ou(w15, clf_gnb_ou),
                 nfl.predict_ou(w16, clf_gnb_ou)]).reset_index(drop = True)

ann = pd.concat([nfl.keras_ou(w2, ann_ou), nfl.keras_ou(w3, ann_ou),
                 nfl.keras_ou(w4, ann_ou), nfl.keras_ou(w5, ann_ou),
                 nfl.keras_ou(w6, ann_ou), nfl.keras_ou(w7, ann_ou),
                 nfl.keras_ou(w8, ann_ou), nfl.keras_ou(w9, ann_ou),
                 nfl.keras_ou(w10, ann_ou), nfl.keras_ou(w11, ann_ou),
                 nfl.keras_ou(w12, ann_ou), nfl.keras_ou(w13, ann_ou),
                 nfl.keras_ou(w14, ann_ou), nfl.keras_ou(w15, ann_ou),
                 nfl.keras_ou(w16, ann_ou)]).reset_index(drop = True)
```

```python
print('RF %:', round(rf.correct.mean()*100, 3))
print('KNN %:', round(knn.correct.mean()*100, 3))
print('LR %:', round(lr.correct.mean()*100, 3))
print('GNB %:', round(gnb.correct.mean()*100, 3))
print('ANN %:', round(ann.correct.mean()*100, 3))
```
